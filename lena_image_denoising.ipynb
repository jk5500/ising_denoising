```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
I=np.array(Image.open('lena_noisy.jpg').convert('L'))
plt.imshow(I,plt.cm.gray)
plt.show()
```

```python
[m,n]=I.shape
Final_image=np.zeros([m,n])
y=I/255.0
x=y
for i in range(1,m-1):
    for j in range(1,n-1):
        #HIDDEN STATE==NEW_VALUE,X
        new_value=np.linspace(0,255,(255+1)) #all possible state we may think to be present
        new_value=new_value/255.0
        Energy=(new_value-y[i,j])**2+((abs(x[i,j+1]-new_value)+abs(x[i,j-1]-new_value)+abs(x[i-1,j]-new_value)+abs(x[i+1,j]-new_value)))
        select=np.argmin(Energy)
        Final_image[i,j]=new_value[select]
```

```python
plt.imshow(Final_image,plt.cm.gray)
plt.title('clean image')
plt.show()
```

```python
I=np.array(Image.open('lena_noisy.jpg').convert('L'))
plt.imshow(I,plt.cm.gray)
plt.show()
plt.imshow(Final_image,plt.cm.gray)
plt.title('clean image')
plt.show()
```
