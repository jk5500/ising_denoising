```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
x=Image.open('in.png').convert('L')
plt.imshow(x,plt.cm.gray)
plt.show()
```

```python
x=np.array(Image.open('in.png').convert('L'))
[m,n]=x.shape
myimage=np.zeros([m,n])
for i in range(m):
    for j in range(n):
        if x[i,j]==255:
            myimage[i,j]=1
        else:
            myimage[i,j]=-1
m,n
```

```python
plt.imshow(myimage,plt.cm.gray)
plt.show()
```

```python
for i in range(m):
    for j in range(n):
        p=np.random.rand()
        if p< 0.1: #bishop 10 percent probability of flipping the sign
            myimage[i,j]=-myimage[i,j]
```

```python
plt.imshow(myimage,plt.cm.gray)
plt.title('noisy image')
plt.show()
```

```python
Final_image=np.zeros([m,n])
y=myimage
x=y
for i in range(1,m-1):
    for j in range(1,n-1):
        old_value=x[i,j]
        new_value=np.array([old_value,-old_value])
        Energy=-2.1*new_value*y[i,j]-(x[i,j+1]+x[i,j-1]+x[i-1,j]+x[i+1,j])*new_value
        select=np.argmin(Energy)
        Final_image[i,j]=new_value[select]
        
        
```

```python
plt.imshow(Final_image,plt.cm.gray)
plt.title('clean image')
plt.show()
```
